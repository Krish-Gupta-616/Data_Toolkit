**4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?**


```python
df1 = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']]
cm = df1.corr()
sns.heatmap(cm, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

**5. Generate a bar plot using Plotly**


```python
import plotly.express as px
avg_age = df.groupby('class')['age'].mean()
gph = px.bar(x=avg_age.index, y=avg_age.values, labels={'x':'Class', 'y':'Average Age'}, title='Average Age by Class')
gph.show()
```

**7. Write a program to perform element-wise multiplication of two NumPy arrays**


```python
x = np.array([1,2,3,4,5])
y = np.array([5,4,3,2,1])
a = np.array([])
for i in range(5):
    a = np.append(a,x[i] * y[i])
a = a.astype(int)
a
```

**8. Create a line plot with multiple lines using Matplotlib**


```python
x = np.linspace(0,3,1000)
y = 5 * x
y1 = x ** 2
y2 = 5*np.sin(5 * x) + 5
y3 = np.exp(x)
plt.plot(x,y, label='y=5x')
plt.plot(x,y1, label='y=x^2')
plt.plot(x,y2, label='y=5sin(5x)+5')
plt.plot(x,y3, label='y=exp(x)')
plt.legend()
plt.title('Multiple Line Plot')
```
